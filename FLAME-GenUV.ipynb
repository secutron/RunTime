```
import argparse

parser = argparse.ArgumentParser(description='Build texture from image')
parser.add_argument('--source_img', default='/content/drive/My Drive/VOCA-flame/hkim정면crop.png', help='source image filename')
parser.add_argument('--target_mesh', default='/content/drive/My Drive/FLAME-result/hkim정면crop.obj', help='target mesh filename')
parser.add_argument('--target_scale', default='/content/drive/My Drive/FLAME-result/hkim정면crop_scale.npy', help='scale of the target mesh for the image projection')
parser.add_argument('--texture_mapping', default='./data/texture_data.npy', help='pre-computed FLAME texture mapping')
parser.add_argument('--out_path', default='/content/drive/My Drive/animation_texture/meshes', help='output path')

args = parser.parse_args(args=[])

# add new path



# Acheive useful directories
source_img_fname = args.source_img
target_mesh_fname = args.target_mesh
target_scale_fname = args.target_scale
texture_mapping = args.texture_mapping
out_path = args.out_path

print(args)
```

```
import os
import cv2
import sys
import argparse
import numpy as np
from psbody.mesh import Mesh
from psbody.mesh.meshviewer import MeshViewer
from utils.project_on_mesh import compute_texture_map

import glob


if not os.path.exists(source_img_fname):
  print('Source image not found - %s' % source_img_fname)
  
if not os.path.exists(target_mesh_fname):
  print('Target mesh not found - %s' % target_mesh_fname)
 
if not os.path.exists(target_scale_fname):
  print('Scale information for target mesh not found %s' % target_scale_fname)

if not os.path.exists(texture_mapping):
  print('Pre-computed FLAME texture mapping not found %s' % texture_mapping)


if not os.path.exists(out_path):
    os.makedirs(out_path)


# Load sequence files
sequence_fnames = sorted(glob.glob(os.path.join('/content/drive/My Drive/animation_output/meshes', '*.obj')))
num_frames = len(sequence_fnames)
if num_frames == 0:
  print('No sequence meshes found')


source_img = cv2.imread(source_img_fname)
tmp_mesh = Mesh(filename="/content/drive/My Drive/animation_output/meshes/00000.obj")

for frame_idx in range(num_frames):
  
  target_mesh = Mesh(filename=sequence_fnames[frame_idx]) #filename=target_mesh_fname)
  # 이거야 이거!
  target_mesh.f = tmp_mesh.f
  #

  target_mesh.set_vertex_colors('white')
  target_scale = np.load(target_scale_fname)

  if sys.version_info >= (3, 0):
      texture_data = np.load(texture_mapping, allow_pickle=True, encoding='latin1').item()
  else:
      texture_data = np.load(texture_mapping, allow_pickle=True).item()
  texture_map = compute_texture_map(source_img, target_mesh, target_scale, texture_data)

  fn = sequence_fnames[frame_idx]
  out_mesh_fname = os.path.join(out_path, os.path.splitext(os.path.basename(fn))[0] + '.obj')
  out_img_fname = os.path.join(out_path, os.path.splitext(os.path.basename(fn))[0] + '.png')

  print(out_img_fname)

  cv2.imwrite(out_img_fname, texture_map)
  target_mesh.vt = texture_data['vt']
  target_mesh.ft = texture_data['ft']
  target_mesh.set_texture_image(out_img_fname)
  target_mesh.write_obj(out_mesh_fname)
  target_mesh.show()

print('*** hkim: done ***')
```
